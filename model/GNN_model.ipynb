## Make recommendations for each user based on rating. GCN



```python
import torch
import torch.nn.functional as F
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from math import sqrt


# Encode user and item IDs
user_encoder = LabelEncoder()
item_encoder = LabelEncoder()

merged_df['user_id'] = user_encoder.fit_transform(merged_df['reviewerID'])
merged_df['item_id'] = item_encoder.fit_transform(merged_df['asin']) + merged_df['user_id'].max() + 1

# Split the DataFrame
train_df, test_df = train_test_split(merged_df, test_size=0.3, random_state=42)

def prepare_data(df):
    edge_index = torch.tensor([df['user_id'].values, df['item_id'].values], dtype=torch.long)
    labels = torch.FloatTensor(df['overall'].values / 5.0)  # Normalize ratings
    return Data(edge_index=edge_index, y=labels)

train_data = prepare_data(train_df)
test_data = prepare_data(test_df)

```

```python
class GCNWithEmbeddings(torch.nn.Module):
    def __init__(self, num_users, num_items, embedding_dim, hidden_channels):
        super().__init__()
        self.user_embedding = torch.nn.Embedding(num_embeddings=num_users, embedding_dim=embedding_dim)
        self.item_embedding = torch.nn.Embedding(num_embeddings=num_items, embedding_dim=embedding_dim)
        self.conv1 = GCNConv(embedding_dim, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, 1)  # Output one value per node

    def forward(self, user_nodes, item_nodes, edge_index):
        user_emb = self.user_embedding(user_nodes)
        item_emb = self.item_embedding(item_nodes - num_users)
        x = torch.cat([user_emb, item_emb], dim=0)
        print("Combined Embedding Shape:", x.shape)

        x = self.conv1(x, edge_index)
        x = F.relu(x)
        print("After Conv1 and ReLU Shape:", x.shape)

        x = self.conv2(x, edge_index)
        print("Final Output Shape:", x.shape)
        user_outputs = x[edge_index[0]]  # Get outputs for user nodes involved in each interaction
        item_outputs = x[edge_index[1]]  # Get outputs for item nodes involved in each interaction

        # You could use some method to combine these outputs, such as addition or multiplication
        # Here's a simple element-wise multiplication
        interaction_outputs = user_outputs * item_outputs

        return interaction_outputs.squeeze()

        #return x.squeeze()


num_users = train_data.edge_index[0].max().item() + 1
num_items = train_data.edge_index[1].max().item() + 1 - num_users
embedding_dim = 16  # You can choose a suitable embedding dimension

model = GCNWithEmbeddings(num_users=num_users, num_items=num_items, embedding_dim=embedding_dim, hidden_channels=16)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = torch.nn.MSELoss()

def train(train_data):
    model.train()
    optimizer.zero_grad()
    out = model(train_data.edge_index[0], train_data.edge_index[1], train_data.edge_index)
    # Compute loss only on the edges, i.e., where we have the actual ratings
    loss = criterion(out[train_data.edge_index[0]], train_data.y)
    loss.backward()
    optimizer.step()
    return loss

# Dummy feature (1s) for each node (user and item)
train_data.x = torch.ones((max(train_data.edge_index.max(), test_data.edge_index.max()) + 1, 1), dtype=torch.float)

losses = []
for epoch in range(100):
    loss = train(train_data)
    losses.append(loss.item())
    print(f'Epoch: {epoch+1:03d}, Loss: {loss.item():.4f}')

plt.plot(range(1, len(losses) + 1), losses, label='Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss over Epochs')
plt.legend()
plt.grid(True)
plt.show()

# Evaluation code would go here - you'll want to test the model on the test_data
# and calculate metrics like RMSE to understand the model performance
```

```python
class GCNWithEmbeddings(torch.nn.Module):
    def __init__(self, num_users, num_items, embedding_dim, hidden_channels):
        super(GCNWithEmbeddings, self).__init__()
        self.user_embedding = torch.nn.Embedding(num_embeddings=num_users, embedding_dim=embedding_dim)
        self.item_embedding = torch.nn.Embedding(num_embeddings=num_items, embedding_dim=embedding_dim)
        self.conv1 = GCNConv(embedding_dim, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, 1)  # Output one value per node (rating prediction)

    def forward(self, user_nodes, item_nodes, edge_index):
        # Get the embeddings for users and items
        user_emb = self.user_embedding(user_nodes)
        item_emb = self.item_embedding(item_nodes)

        # Concatenate the user and item embeddings to form the node features
        x = torch.cat([user_emb, item_emb], dim=0)

        # Apply the GCN layers
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        return x.squeeze()
```

```python
import torch
from sklearn.preprocessing import LabelEncoder

# Assume merged_df has columns: 'reviewerID', 'asin' (item ID), 'overall' (rating)
user_encoder = LabelEncoder()
item_encoder = LabelEncoder()

# Encode user and item IDs
merged_df['user_id'] = user_encoder.fit_transform(merged_df['reviewerID'])
merged_df['item_id'] = item_encoder.fit_transform(merged_df['asin']) + merged_df['user_id'].max() + 1

# Create edge_index and edge_attr for GNN input
edge_index = torch.tensor([merged_df['user_id'].values, merged_df['item_id'].values], dtype=torch.long)
edge_attr = torch.tensor(merged_df['overall'].values, dtype=torch.float).unsqueeze(1)  # Ratings as edge attributes

# Dummy node features (one-hot encoding could be a starting point for actual features)
num_nodes = max(merged_df['item_id'].max(), merged_df['user_id'].max()) + 1
x = torch.eye(num_nodes)  # Identity matrix as dummy features
```

```python
from torch_geometric.nn import GCNConv, global_mean_pool
import torch.nn.functional as F
import torch.nn as nn

class GNN(torch.nn.Module):
    def __init__(self, num_features, hidden_channels):
        super(GNN, self).__init__()
        self.conv1 = GCNConv(num_features, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, hidden_channels)
        self.out = nn.Linear(hidden_channels, 1)  # Output one value per node (rating prediction)

    def forward(self, x, edge_index, batch):
        # Initial node features are passed through GNN layers
        x = F.relu(self.conv1(x, edge_index))
        x = F.relu(self.conv2(x, edge_index))
        x = self.out(x)
        return x
```

```python
from torch_geometric.data import Data

data = Data(x=x, edge_index=edge_index, edge_attr=edge_attr)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model = GNN(num_features=x.size(1), hidden_channels=64).to(device)
data = data.to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = torch.nn.MSELoss()

for epoch in range(200):  # Number of epochs
    model.train()
    optimizer.zero_grad()
    out = model(data.x, data.edge_index, torch.zeros(data.num_nodes, dtype=torch.long).to(device))
    loss = criterion(out.squeeze(), data.edge_attr)
    loss.backward()
    optimizer.step()
    print(f'Epoch {epoch}, Loss: {loss.item()}')
```

```python
model.eval()
with torch.no_grad():
    predictions = model(data.x, data.edge_index, torch.zeros(data.num_nodes, dtype=torch.long).to(device))
```

```python
from bs4 import BeautifulSoup
import re
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

# Download NLTK stopwords
nltk.download('stopwords')
# Combine NLTK and sklearn's stopword lists for a more comprehensive set
stop_words = set(stopwords.words('english')).union(set(ENGLISH_STOP_WORDS))

def clean_text(text):
    # Remove HTML tags
    soup = BeautifulSoup(text, "html.parser")
    text = soup.get_text(separator=" ")
    # Remove special characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    # Remove stop words
    text = " ".join([word for word in text.split() if word.lower() not in stop_words])
    return text

# Assuming 'df' is your DataFrame and it has a 'description' column with HTML content
filtered_df['cleaned_description'] = filtered_df['description'].apply(lambda x: ' '.join(x)).apply(clean_text)
```

```python
from flask import Flask, render_template, request
import pandas as pd
import plotly.graph_objects as go
import networkx as nx
import plotly.graph_objects as go

# Initialize Flask app
app = Flask(__name__)


# Define plot_node_plot function
import networkx as nx
import plotly.graph_objects as go

def plot_node_plot(asin, filtered_df):
    # Find top similar items for the given ASIN
    top_similar_items = filtered_df.loc[filtered_df['asin'] == asin, 'Top_Similar_Items'].iloc[0]

    # Create a directed graph
    G = nx.DiGraph()

    # Add nodes (ASIN and similar items)
    G.add_node(asin, color='blue', label=asin)
    for item in top_similar_items:
        G.add_node(item, color='red', label=item)

    # Add edges from ASIN to similar items
    for item in top_similar_items:
        G.add_edge(asin, item)

    # Create node positions
    pos = nx.spring_layout(G)

    # Extract node positions for Plotly
    node_x = [pos[node][0] for node in G.nodes()]
    node_y = [pos[node][1] for node in G.nodes()]

    # Create edges
    edge_x = []
    edge_y = []
    for edge in G.edges():
        x0, y0 = pos[edge[0]]
        x1, y1 = pos[edge[1]]
        edge_x.extend([x0, x1, None])
        edge_y.extend([y0, y1, None])

    # Create node trace
    node_trace = go.Scatter(
        x=node_x, y=node_y,
        mode='markers+text',
        hoverinfo='text',
        marker=dict(
            size=20,
            color=[G.nodes[n]['color'] for n in G.nodes()]
        ),
        text=[G.nodes[n]['label'] for n in G.nodes()],
        textposition="top center"
    )

    # Create edge trace
    edge_trace = go.Scatter(
        x=edge_x, y=edge_y,
        line=dict(width=1, color='#888'),
        hoverinfo='none',
        mode='lines'
    )

    # Create figure
    fig = go.Figure(data=[edge_trace, node_trace],
                    layout=go.Layout(
                        title=f'Node Plot for ASIN {asin} with Top Similar Items',
                        titlefont_size=16,
                        showlegend=False,
                        hovermode='closest',
                        margin=dict(b=20, l=5, r=5, t=40),
                        xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                        yaxis=dict(showgrid=False, zeroline=False, showticklabels=False))
                    )

    # Show plot
    fig.show()


# Define route for home page
@app.route('/')
def home():
    return render_template('/content/drive/MyDrive/Social_Network/Xueyuan_test/templates/index.html')

# Define route to handle form submission
@app.route('/recommendation', methods=['POST'])
def recommendation():
    asin_input = request.form['asin']
    plot_node_plot(asin_input, filtered_df)
    return render_template('/content/drive/MyDrive/Social_Network/Xueyuan_test/templates/recommendation.html')

# Run Flask app
if __name__ == '__main__':
    app.run(debug=True)
```
