```python
import torch
import torch.nn.functional as F
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from math import sqrt


# Encode user and item IDs
user_encoder = LabelEncoder()
item_encoder = LabelEncoder()

merged_df['user_id'] = user_encoder.fit_transform(merged_df['reviewerID'])
merged_df['item_id'] = item_encoder.fit_transform(merged_df['asin']) + merged_df['user_id'].max() + 1

# Split the DataFrame
train_df, test_df = train_test_split(merged_df, test_size=0.3, random_state=42)

def prepare_data(df):
    edge_index = torch.tensor([df['user_id'].values, df['item_id'].values], dtype=torch.long)
    labels = torch.FloatTensor(df['overall'].values / 5.0)  # Normalize ratings
    return Data(edge_index=edge_index, y=labels)

train_data = prepare_data(train_df)
test_data = prepare_data(test_df)
```

```python
class GCNWithEmbeddings(torch.nn.Module):
    def __init__(self, num_users, num_items, embedding_dim, hidden_channels):
        super().__init__()
        self.user_embedding = torch.nn.Embedding(num_embeddings=num_users, embedding_dim=embedding_dim)
        self.item_embedding = torch.nn.Embedding(num_embeddings=num_items, embedding_dim=embedding_dim)
        self.conv1 = GCNConv(embedding_dim, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, 1)  # Output one value per node

    def forward(self, user_nodes, item_nodes, edge_index):
        user_emb = self.user_embedding(user_nodes)
        item_emb = self.item_embedding(item_nodes - num_users)
        x = torch.cat([user_emb, item_emb], dim=0)
        print("Combined Embedding Shape:", x.shape)

        x = self.conv1(x, edge_index)
        x = F.relu(x)
        print("After Conv1 and ReLU Shape:", x.shape)

        x = self.conv2(x, edge_index)
        print("Final Output Shape:", x.shape)
        user_outputs = x[edge_index[0]]  # Get outputs for user nodes involved in each interaction
        item_outputs = x[edge_index[1]]  # Get outputs for item nodes involved in each interaction

        # You could use some method to combine these outputs, such as addition or multiplication
        # Here's a simple element-wise multiplication
        interaction_outputs = user_outputs * item_outputs

        return interaction_outputs.squeeze()

        #return x.squeeze()


num_users = train_data.edge_index[0].max().item() + 1
num_items = train_data.edge_index[1].max().item() + 1 - num_users
embedding_dim = 16  # You can choose a suitable embedding dimension

model = GCNWithEmbeddings(num_users=num_users, num_items=num_items, embedding_dim=embedding_dim, hidden_channels=16)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = torch.nn.MSELoss()

def train(train_data):
    model.train()
    optimizer.zero_grad()
    out = model(train_data.edge_index[0], train_data.edge_index[1], train_data.edge_index)
    # Compute loss only on the edges, i.e., where we have the actual ratings
    loss = criterion(out[train_data.edge_index[0]], train_data.y)
    loss.backward()
    optimizer.step()
    return loss

# Dummy feature (1s) for each node (user and item)
train_data.x = torch.ones((max(train_data.edge_index.max(), test_data.edge_index.max()) + 1, 1), dtype=torch.float)
```

1. Feature Encoding and Preparation
For text-based features like category and description, you typically need to convert these into numerical format that a model can process. Common techniques include:

- One-hot encoding for categorical features like category.
- TF-IDF or Word Embeddings for textual descriptions.

2. Integrating Features into the GNN
- Node Features: Incorporate the encoded features into the node representations in your graph. Each node will have features that might include the encoded category and text description embeddings.
- Graph Structure: Maintain the user-item graph structure where edges represent interactions (ratings), and nodes represent users and items.

```python
import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv

class GCNWithFeatures(torch.nn.Module):
    def __init__(self, num_users, num_items, user_features_dim, item_features_dim, hidden_channels):
        super().__init__()
        self.user_features_embedding = torch.nn.Linear(user_features_dim, hidden_channels)
        self.item_features_embedding = torch.nn.Linear(item_features_dim, hidden_channels)
        self.conv1 = GCNConv(hidden_channels, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, 1)  # Output one value per node

    def forward(self, user_features, item_features, edge_index):
        # Embed features
        user_features = F.relu(self.user_features_embedding(user_features))
        item_features = F.relu(self.item_features_embedding(item_features))
        x = torch.cat([user_features, item_features], dim=0)

        # Graph convolution layers
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        return x.squeeze()
```

```python
import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv, global_mean_pool
from torch.nn import Linear, Embedding

class RecommenderSystem(torch.nn.Module):
    def __init__(self, num_users, num_items, user_features_dim, item_features_dim, hidden_dim):
        super(RecommenderSystem, self).__init__()
        self.user_embedding = Embedding(num_users, embedding_dim=hidden_dim)
        self.item_embedding = Embedding(num_items, embedding_dim=hidden_dim)

        # GCN Layers
        self.conv1 = GCNConv(hidden_dim + user_features_dim + item_features_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, hidden_dim)

        # Final output layer
        self.out = Linear(hidden_dim, 1)

    def forward(self, x, edge_index, edge_type, batch):
        # Apply embeddings
        user_embed = self.user_embedding(edge_type[0])
        item_embed = self.item_embedding(edge_type[1] - num_users)

        # Concatenate embeddings with features
        x = torch.cat([x, user_embed, item_embed], dim=1)

        # First GCN Layer
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=0.5, training=self.training)

        # Second GCN Layer
        x = self.conv2(x, edge_index)
        x = F.relu(x)

        # Global mean pooling
        x = global_mean_pool(x, batch)  # Aggregate features

        # Output layer
        x = self.out(x)
        return x.squeeze()
```

```python
import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch.nn import Linear, Embedding

class RecommenderSystem(torch.nn.Module):
    def __init__(self, num_users, num_items, feature_dim, hidden_dim):
        super(RecommenderSystem, self).__init__()
        self.num_users = num_users
        self.num_items = num_items
        self.user_embedding = Embedding(num_users, embedding_dim=hidden_dim)
        self.item_embedding = Embedding(num_items, embedding_dim=hidden_dim)

        # GCN Layers - adjusting the input feature dimension
        self.conv1 = GCNConv(hidden_dim * 2 + feature_dim, hidden_dim)  # Note the adjusted dimension here
        self.conv2 = GCNConv(hidden_dim, hidden_dim)

        # Final output layer
        self.out = Linear(hidden_dim, 1)

    def forward(self, x, edge_index):
        user_indices = edge_index[0]
        item_indices = edge_index[1] - self.num_users  # Adjust item indices to start at 0

        user_embed = self.user_embedding(user_indices)
        item_embed = self.item_embedding(item_indices)

        # Concatenate node features with user and item embeddings
        x = torch.cat([x, user_embed, item_embed], dim=1)

        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=0.5, training=self.training)
        x = self.conv2(x, edge_index)
        x = F.relu(x)

        return self.out(x)



# Train and evaluate
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = RecommenderSystem(num_users=merged_df['reviewerID_encoded'].nunique(), num_items=merged_df['item_id_encoded'].nunique(), feature_dim=node_features.shape[1], hidden_dim=64).to(device)
data = data.to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = torch.nn.MSELoss()

def train():
    model.train()
    optimizer.zero_grad()
    output = model(data.x, data.edge_index)
    loss = criterion(output, data.y)
    loss.backward()
    optimizer.step()
    return loss.item()

for epoch in range(100):  # Number of epochs
    loss = train()
    print(f'Epoch {epoch}: Loss {loss:.4f}')

# Predicting ratings
def predict():
    model.eval()
    with torch.no_grad():
        predictions = model(data.x, data.edge_index)
    return predictions

# Get predictions
predicted_ratings = predict().cpu().numpy()
merged_df['pred_rating'] = predicted_ratings

# Now, merged_df includes the predicted ratings
print(merged_df[['asin', 'reviewerID', 'overall', 'pred_rating']])
```
